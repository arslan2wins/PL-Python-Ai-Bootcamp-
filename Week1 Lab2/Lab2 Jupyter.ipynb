```python
text = "Python is simple yet powerful. Python is easy to learn."
text = text.replace(".", "")
text = text.replace(",", "")
text = text.lower()
words = text.split()
print("Words list:", words)
word_freq = {}
for word in words:
    if word in word_freq:
        word_freq[word] += 1 
    else:
        word_freq[word] = 1
print("Word Frequencies", word_freq)
```
