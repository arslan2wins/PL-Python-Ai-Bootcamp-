```python
import numpy as np
temp = df['temperature_C'].to_numpy()
wind = df['wind_speed_kmph'].to_numpy()
```

```python
temp_matrix = temp.reshape((10, 24))

daily_min = temp_matrix.min(axis=1)
daily_max = temp_matrix.max(axis=1)
daily_mean = temp_matrix.mean(axis=1)

print("Daily Min:", daily_min)
print("Daily Max:", daily_max)
print("Daily Mean:", daily_mean)
```

```python
def normalize(matrix):
    return (matrix - matrix.mean()) / matrix.std()

normalized_temp_matrix = normalize(temp_matrix)
print(normalized_temp_matrix)
```

```python
mask = wind > 15
high_wind = wind[mask]

print("Total readings:", len(wind))
print("High wind readings (>15 kmph):", len(high_wind))
print("High wind values:", high_wind)
```

```python
def daily_summary(matrix):
    summaries = []                          
    
    for i in range(len(matrix)):            
        day_data = matrix[i]                
        
        summary = {
            "day": i + 1,
            "min": day_data.min(),
            "max": day_data.max(),
            "mean": day_data.mean()
        }
        
        summaries.append(summary)           
    
    return summaries                        
```

```python
summaries = daily_summary(temp_matrix)

for day in summaries:
    print(day)
```
