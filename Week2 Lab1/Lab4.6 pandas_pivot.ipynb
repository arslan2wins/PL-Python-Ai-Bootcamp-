<h1 style="color:blue">Pivot Table</h1>

```python
df = pd.read_csv("/content/drive/MyDrive/pandas_pivot/weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date")
#agar 2 same values values hon gi to pivot uska mean  nikal de ga automaticaly, jesa k 2 cities hen or dates b 2 hen
```

<h2 style="color:brown">Margins</h2>

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc="sum")
#margins ek extra row aur extra column add kar deta hai jiska naam "All" hota hai — jo totals dikhata hai
#aggfunc is average function, pehle ye mean tha lekin ab btaya hai k mean nhi sum nikalo
```

<h2 style="color:brown">Grouper</h2>

```python
df = pd.read_csv("/content/drive/MyDrive/pandas_pivot/weather3.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```

```python
df.groupby(pd.Grouper(key='date', freq='Y'))['temperature'].mean()
```

```python
df.groupby(['city', pd.Grouper(key='date', freq='M')])['temperature'].mean()
```
